```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Agregar el archivo: VentasTotales.csv
data = pd.read_csv('Ventas_totales.csv')
data.head(10)
```

```python
#Identificar valores nulos por columna
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#Ver tipo de datos
data.info()
```

Sustitucion de valores nulos

```python
#Realizamos una copia del dataframe
data2=data.copy()
```

```python
#Sustitucion columna salon_ventas 
#Sustituir valores nulos con promedio o media ya que las ventas del salón son varian muy poco
data2["salon_ventas"]=data2["salon_ventas"].fillna(round(data["salon_ventas"].mean(),1))
```

```python
#Corroboramos valores nulos col salon_ventas 
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustitucion columna tarjetas_debito
#Sustituir valores nulos con la moda ya que este atributo 
# representa la cantidad que mas se repite de tarjetas de debito a lo largo del tiempo
data2["tarjetas_debito"]=data2["tarjetas_debito"].fillna(round(data["tarjetas_debito"].mode(),1))
```

```python
#Corroboramos valores nulos col tarjetas_debito
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustitucion columna tarjetas_credito
#Sustituir valores nulos con la moda ya que este atributo 
# representa la cantidad que mas se repite de tarjetas de crédito a lo largo del tiempo
data2["tarjetas_credito"]=data2["tarjetas_credito"].fillna(round(data["tarjetas_credito"].mode(),1))
```

```python
#Corroboramos valores nulos col tarjetas_credito
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustitucion columna otros_medios
#Sustituir valores nulos con promedio o media ya que es un dato que balancea la información del dato faltante de esta columna
data2["otros_medios"]=data2["otros_medios"].fillna(round(data["otros_medios"].mean(),1))
```

```python
#Corroboramos valores nulos col otros_medios
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustitucion columna subtotal_ventas_alimentos_bebidas
#Sustituir valores nulos con media ya que al tratarse de $ el promedio es lo necesario para sustituir los datos nulos de la columna
data2["subtotal_ventas_alimentos_bebidas"]=data2["subtotal_ventas_alimentos_bebidas"].fillna(round(data["subtotal_ventas_alimentos_bebidas"].mean(),1))
data2
```

```python
#Corroboramos valores nulos col subtotal_ventas_alimentos_bebidas
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustitucion columna bebidas
#colBebidas=data.iloc[:,[13]]
#colBebidas
#Sustituir valores nulos con la moda ya que este atributo 
# representa la cantidad que más se venden de bebidas
data2["bebidas"]=data2["bebidas"].fillna(round(data["bebidas"].mode(),1))
```

```python
#Corroboramos valores nulos col bebidas
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustitucion columna almacen
#Sustituir valores nulos con media ya que es el promedio de productos de almacen que tienen
data2["almacen"]=data2["almacen"].fillna(round(data["almacen"].mean(),1))
data2
```

```python
#Corroboramos valores nulos col almacen
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustitucion columna panaderia
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill") ya que
# la cantidad de pedidos por pan varia por día, entonces para que el numero se adapte al del mismo día, día anterior o día despues
data2["panaderia"] =data2["panaderia"].fillna(method="ffill") 
```

```python
#Corroboramos valores nulos col panaderia
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustitucion columna lacteos
# Con promedio ya que los valores no tienen tanto rango
data2["lacteos"]=data2["lacteos"].fillna(round(data["lacteos"].mean(),1))
data2
```

```python
#Corroboramos valores nulos col lacteos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustitucion columna carnes
# Con promedio ya que los valores no tienen tanto rango
data2["carnes"]=data2["carnes"].fillna(round(data["carnes"].mean(),1))
data2
```

```python
#Corroboramos valores nulos col carnes
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustitucion columna verduleria_fruteria
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill") ya que
# la cantidad de pedidos por pan varia por día, entonces para que el numero se adapte al del mismo día, día anterior o día despues
data2["verduleria_fruteria"] =data2["verduleria_fruteria"].fillna(method="ffill") 
```

```python
#Corroboramos valores nulos col lacteos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustitucion columna alimentos_preparados_rotiseria
#Sustituir valores nulos con la moda ya que este atributo 
# representa la cantidad que mas se repite de alimentos que se preparan en la rostiseria
data2["alimentos_preparados_rotiseria"]=data2["alimentos_preparados_rotiseria"].fillna(round(data["alimentos_preparados_rotiseria"].mode(),1))
```

```python
#Corroboramos valores nulos col lacteos
valores_nulos=data2.isnull().sum()
valores_nulos
```
